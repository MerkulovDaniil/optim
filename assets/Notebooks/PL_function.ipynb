```python
import plotly.graph_objects as go
import numpy as np
import plotly.io as pio

pio.renderers.default = "plotly_mimetype+notebook_connected"

# Create a meshgrid for (x, y) values
x = np.linspace(-5, 5, 25)
y = np.linspace(-2, 2, 25)
x, y = np.meshgrid(x, y)

# Compute the function values based on the meshgrid
z = (y - np.sin(x)) ** 2 / 2

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='Sunset')])

# Add axis labels and title
fig.update_layout(scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='f(x,y)'
    ),
    title='Non-convex PL function f(x, y) = (y - sin(x))/2',
    template="plotly_white"
)

plotly_config = {
    'displaylogo': False,
    'toImageButtonOptions': {
        'format': 'svg',
        'filename': 'fmin',
        'height': None,
        'width': None,
        'scale': 1
    },
    'modeBarButtonsToRemove': ['select2d', 'lasso2d'],
    'modeBarButtonsToAdd': ['drawopenpath', 'eraseshape']
}

fig.write_html(
        "PL_non_convex.html",
        config=plotly_config,
        include_plotlyjs="cdn",
        full_html=False,
    )

fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return x**2 + 3*np.sin(x)**2

# Generate x values
x = np.linspace(-3, 3, 400)

# Generate y values based on the function
y = f(x)

# Create the plot
plt.figure(figsize=(3, 3))
plt.plot(x, y, linewidth=2, label=r"$f(x) = x^2 + 3sin^2(x)$")
plt.grid(linestyle=":")

# Customizing the ticks
plt.xticks(np.arange(-3, 4, 1))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()

# Save the figure as a PDF
plt.title("Function, that satisfies\n Polyak- Lojasiewicz condition")
plt.tight_layout()
plt.savefig("pl_2d.pdf", format='pdf')

# Show the plot for visualization
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x, y):
    return (y - np.sin(x)) ** 2 / 2

# Generate x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-2, 2, 100)

# Create meshgrid
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create the 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')


# Plot the surface
surf = ax.plot_surface(X, Y, Z, alpha=0.8, cmap='coolwarm', edgecolor='none')

# Customizing the ticks and labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Non-convex PL function')
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
ax.xaxis.pane.set_edgecolor('w')
ax.yaxis.pane.set_edgecolor('w')
ax.zaxis.pane.set_edgecolor('w')
ax.xaxis._axinfo["grid"]['linestyle'] = ":"
ax.yaxis._axinfo["grid"]['linestyle'] = ":"
ax.zaxis._axinfo["grid"]['linestyle'] = ":"

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.tight_layout()
plt.savefig("pl_3d.pdf", format='pdf')
plt.show()
```
